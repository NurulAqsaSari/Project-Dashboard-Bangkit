```python
#melihat rata-rata jumlah peminjaman sepeda terhadap keadaan cuaca dan musim

datab = day.copy()

# Melakukan mapping untuk setiap variabel kategorikal
datab['weathersit'] = datab['weathersit'].map({1: 'Cerah', 2: 'Berkabut', 3: 'Hujan Ringan', 4: 'Hujan Lebat'})
datab['season'] = datab['season'].map({1: 'springer', 2: 'summer', 3: 'fall', 4: 'winter'})

# Mengatur urutan kelas
weathersit_order = ['Cerah', 'Berkabut', 'Hujan Ringan', 'Hujan Lebat']
season_order = ['springer', 'summer', 'fall', 'winter']

# Membuat pivot tabel untuk rata-rata jumlah peminjaman sepeda terhadap variabel cuaca dan musim
pivot_weathersit = datab.groupby(by=["weathersit"]).agg({
    "cnt": "mean"
}).reindex(weathersit_order, fill_value=0).reset_index()
pivot_season = datab.groupby(by=["season"]).agg({
    "cnt": "mean"
}).reindex(season_order, fill_value=0).reset_index()

# Menampilkan pivot tabel
pd.options.display.float_format = '{:.2f}'.format
print("Rata-rata jumlah peminjaman sepeda terhadap keadaan cuaca :")
print(pivot_weathersit.to_string(index=False))
print("\n===============================================\n")
print("Rata-rata jumlah peminjaman sepeda terhadap musim :")
print(pivot_season.to_string(index=False))
```

## Visualization & Explanatory Analysis

- Bagaimana hubungan antara musim dan jumlah sepeda harian?
- Apakah ada pola berdasarkan bulan dalam jumlah sepeda harian?
- Bagaimana pengaruh cuaca (weathersit) terhadap jumlah sewa sepeda harian?

### Pertanyaan 1:

```python
#Hubungan antara musim dan jumlah sewa sepeda harian

seasonal_data = day.groupby('season')['cnt'].mean()
season_names = ['springer', 'summer', 'fall', 'winter']
colors = ["pink", "pink", "darkcyan", "pink"]
sns.barplot(season_names, seasonal_data, palette = colors)
plt.xlabel('Musim')
plt.ylabel('Rata-rata Jumlah Sewa Sepeda Harian')
plt.title('Pengaruh Musim terhadap Jumlah Sewa Sepeda Harian')
plt.show()
```

Berdasarkan grafik diatas dapat dilihat bahwa jumlah sewa sepeda lebih banyak pada musim gugur(Fall)

### Pertanyaan 2:

```python
#pola berdasarkan bulan dalam jumlah sewa sepeda harian

sns.set_style("darkgrid")
plt.figure(figsize=(12, 6))
sns.lineplot(x="mnth", y="cnt", data=day, ci=None, color = "purple")
plt.title("Pola Jumlah Sewa Sepeda Harian Berdasarkan Bulan")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Sewa Sepeda Harian")
plt.show()
```

berdasarkan line chart di atas dapat disimpulkan bahwa dari bulan jumlah sewa sepeda meningkat pada bulan 6 dan 9.

### Pertanyaan 3:

```python
#pengaruh cuaca (weathersit) terhadap jumlah sewa sepeda harian

plt.figure(figsize=(10, 6))
sns.boxplot(x="weathersit", y="cnt", data=day)
plt.title("Pengaruh Weathersit Terhadap Jumlah Sewa Sepeda Harian")
plt.xlabel("Weathersit")
plt.ylabel("Jumlah Sewa Sepeda Harian")
plt.show()
```

Berdasarkan boxplot di atas dapat disimpulkan bahwa jumlah sewa sepeda mengingkat ketika cuaca Cerah, Sedikit awan, Berawan sebagian, Berawan sebagian.
